### Import dataset & Interpolation

```r
iWUE <- read.rwl(fname='./Data/Tree_Ring/iWUE_all_sites.csv',format="csv")
uWUE <- read.rwl(fname='./Data/Tree_Ring/uWUE_all_sites.csv',format="csv")
WUE  <- read.rwl(fname='./Data/Tree_Ring/WUE_all_sites.csv',format="csv")

# fill the internal NA values using cubic spline interpolation
iWUE.fill <- fill.internal.NA(iWUE,fill="Spline")
uWUE.fill <- fill.internal.NA(uWUE,fill="Spline")
WUE.fill  <- fill.internal.NA(WUE,fill="Spline")
```

### calculate 50-year high-pass filtered value for a single site

```r
start_yr = 1900
series <- iWUE.fill$CAZ#COL
yrs    <- seq_along(series)+start_yr

yrs_noNA    <- yrs[!is.na(series)]
series_noNA <- series[!is.na(series)]

spl50 <- ffcsaps(series_noNA,yrs_noNA,nyrs=50,f=0.5)
spl25 <- ffcsaps(series_noNA,yrs_noNA,nyrs=25,f=0.5)
spl75 <- ffcsaps(series_noNA,yrs_noNA,nyrs=75,f=0.5)
```

```r
# jpeg('./Data/figs/smooth_spline.jpg',res=300)
# tiff("./Data/figs/smooth_spline.jpg", units="in", width=5, height=5, res=300)

my.cols <- c("#1B9E77", "#D95F02", "#7570B3")

ylabel <- expression(paste("iWUE (", mu, "mol/mol)"))

plot(yrs,series, type = "l", xlab='Year',ylab = ylabel, col = "grey")
grid(col="black",lwd=0.5)

lines(yrs_noNA,spl75,col=my.cols[1], lwd = 2)
lines(yrs_noNA,spl50,col=my.cols[2],lwd=2)
lines(yrs_noNA,spl25,col=my.cols[3],lwd=2)

legend("topleft", c("original data", "75 yrs", "50 yrs", "25 yrs"), 
       lwd = 2, col = c("grey",my.cols),bg = "white")

title('Site: CAZ')

# dev.off()
```

### calculate 50-year high-pass filtered value for all sites

```r
Conifer   = c('COL','CAZ','SER','PED','LIL','VIG','POE',
             'NIE2','WIN','SUW','PAN','GUT','ILO','INA')
Broadleaf = c('CAV','LAI','REN','FON','NIE1','DRA','WOB','LCH','BRO')
All_sites = c(Conifer , Broadleaf)
```

iWUE

```r
smooth   <- data.frame(year=seq(1901,2010))
residual <- data.frame(year=seq(1901,2010))

for (site in All_sites ){
    series <- iWUE.fill[,site]
    yrs    <- seq(1901,2010)
    
    # remove NA values
    yrs_noNA    <- yrs[!is.na(series)]
    series_noNA <- series[!is.na(series)]
    # 50-year smoothing spline
    spl50 <- ffcsaps(series_noNA,yrs_noNA,nyrs=50,f=0.5)
    
    # restore the NA values 
    new_col <- copy(series)
    new_col[!is.na(new_col)] <- spl50
    
    # put the complete smoothing results to "smooth" dataframe
    smooth[site] <- new_col
    # store the residuals
    residual[site] <- series - new_col
#     break
}
```

```r
write.csv(smooth,file="./Data/Tree_Ring/iWUE_smooth_all_sites.csv")
write.csv(residual,file="./Data/Tree_Ring/iWUE_residual_all_sites.csv")
```
