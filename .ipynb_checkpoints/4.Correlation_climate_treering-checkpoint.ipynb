### import data

```python
Tmp_all = pd.read_csv('./Data/CRU/Temp_monthly_all_sites.csv',index_col=0)
VPD_all = pd.read_csv('./Data/CRU/VPD_monthly_all_sites.csv',index_col=0)
RH_all  = pd.read_csv('./Data/CRU/RH_monthly_all_sites.csv',index_col=0)
pet_all = pd.read_csv('./Data/CRU/Pet_monthly_all_sites.csv',index_col=0)
tmx_all = pd.read_csv('./Data/CRU/Tmx_monthly_all_sites.csv',index_col=0)
pre_all = pd.read_csv('./Data/CRU/Pre_monthly_all_sites.csv',index_col=0)

Tmp_all.index = pd.to_datetime(Tmp_all.index)
VPD_all.index = pd.to_datetime(VPD_all.index)
RH_all.index  = pd.to_datetime(RH_all.index)
pet_all.index = pd.to_datetime(pet_all.index)
tmx_all.index = pd.to_datetime(tmx_all.index)
pre_all.index = pd.to_datetime(pre_all.index)


# select 1901-2010 period
Tmp_all = Tmp_all[Tmp_all.index.year<=2010] 
VPD_all = VPD_all[VPD_all.index.year<=2010] 
RH_all  = RH_all[RH_all.index.year<=2010] 
pet_all = pet_all[pet_all.index.year<=2010] 
tmx_all = tmx_all[tmx_all.index.year<=2010] 
pre_all = pre_all[pre_all.index.year<=2010] 
```

```python
iWUE_rsd = pd.read_csv('./Data/Tree_Ring/iWUE_residual_all_sites.csv',index_col=0)
uWUE_rsd = pd.read_csv('./Data/Tree_Ring/uWUE_residual_all_sites.csv',index_col=0)
WUE_rsd  = pd.read_csv('./Data/Tree_Ring/WUE_residual_all_sites.csv',index_col=0)

iWUE = pd.read_csv('./Data/Tree_Ring/iWUE_all_sites.csv',index_col=0)
uWUE = pd.read_csv('./Data/Tree_Ring/uWUE_all_sites.csv',index_col=0)
WUE  = pd.read_csv('./Data/Tree_Ring/WUE_all_sites.csv',index_col=0)

iWUE_rsd = iWUE_rsd.set_index('year')
uWUE_rsd = uWUE_rsd.set_index('year')
WUE_rsd  =  WUE_rsd.set_index('year')
```

```python
All_sites = ['COL','CAZ','SER','PED','LIL','CAV','VIG','POE','LAI','REN','FON',
             'NIE1','NIE2','WIN','DRA','WOB','SUW','PAN','LCH','BRO','GUT','ILO','INA']
```

### correlations for each month

```python
R_dic = {}
p_dic = {}
var_s = ['VPD','Tmp','RH','pet','tmx','pre']

for (j,var_all) in enumerate([VPD_all,Tmp_all,RH_all,pet_all,tmx_all,pre_all]):
    R_list = []
    p_list = []
    for site in All_sites:
        R_site = []
        p_site = []
        for i in range(1,13):

            var_month = var_all[var_all.index.month==i]
            var_month = var_month.set_index(iWUE_rsd.index)

            # remove NAN for following regression
            mask = (var_month[site].isna()==False)&(iWUE_rsd[site].isna()==False)
            x    = var_month[site][mask]
            y    = iWUE_rsd[site][mask]

            slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
            R_site.append(r_value)
            p_site.append(p_value)
        R_list.append(R_site)
        p_list.append(p_site)
        
    R_list = np.array(R_list)
    p_list = np.array(p_list)
    
    R_dic[var_s[j]] = R_list
    p_dic[var_s[j]] = p_list
```

### VPD

```python
# set changing colors for different months
Ncolors   = 23
colormap  = plt.cm.Spectral
mapcolors = [colormap(int(x*colormap.N/Ncolors)) for x in range(Ncolors)]
```

```python
plt.rcParams['font.size'] = 11
R_list = R_dic['VPD']
fig, ax  = plt.subplots(figsize=(16,6))
ax.set_position([0.1,0.1,0.5,0.8])
barWidth = 0.05
xtickLabel = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

r = np.arange(12)

for (i,R_site) in enumerate(R_list):
    
    
    ax.bar(r, R_site, color=mapcolors[22-i],width=barWidth, edgecolor='white', label=All_sites[i])
    
    r = [x + barWidth for x in r]
#     break
ax.set_xticks(np.arange(12))
ax.set_xticklabels(xtickLabel,fontsize=12,fontweight='bold')
ax.set_ylabel('Correlation Coefficient with VPD (r)',fontsize=12,fontweight='bold')
ax.set_ylim([-0.45,0.6])
ax.set_xlim([-1,13])

xmin, xmax = ax.get_xlim()
ci = ax.hlines(y=[-0.195,0.195],xmin=xmin,xmax=xmax, color="k",linewidths=1,linestyles='--',label='95% \nconfidence \nlevel')

ax.legend(bbox_to_anchor=(1.0, 1.05),frameon=False)

plt.rcParams['savefig.dpi'] = 220
T = dt.datetime.now().strftime("%Y%m%d-%H%M%S")
# plt.savefig('./Data/figs/All_sites_VPD_correlation_'+T+'.png')
```
